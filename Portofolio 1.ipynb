```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file = 'train.csv'
loadcsv = pd.read_csv(file)
```

```python
display(loadcsv)
```

```python
def heatmap(df):
    plt.figure(figsize=(20,20))
    sns.heatmap(df.corr(), annot=True, cmap='coolwarm',linewidths=0.5)
    plt.show()
    
heatmap(loadcsv)
```

```python
for kolom in loadcsv.columns:
    plt.hist(loadcsv[kolom], bins=10, alpha=0.7, edgecolor='black')
    plt.title(f'Histogram van {kolom}')
    plt.xlabel(kolom)
    plt.show()
```

```python
loadcsv.head()
```

```python
def searchrow(df, column, value):
    return df[df[column] == value]
searchrow(loadcsv, 'bmi', 97.6)
```

```python
def filterdf(df, column, value):
    return df[df[column] == value]
stroke_df = filterdf(loadcsv, 'stroke', 1)
```

```python
def scatterplot(df, x, y, z):
    sns.scatterplot(df, x=x, y=y, hue=z)
    plt.title(f'{x} vs {y}')
    plt.xlabel(x)
    plt.ylabel(y)
    plt.show()
    
scatterplot(stroke_df, 'age', 'bmi', 'smoking_status_formerly smoked')
```
